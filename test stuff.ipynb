```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([[1],[2],[3],[4]])
y = np.array([[1.1],[1.9],[3.1],[3.9]])

x_matrix = np.hstack([x, x*x, x*x*x])  


# Use the sci-kit learn stuff
linreg3 = LinearRegression()
linreg3.fit(x_matrix, y)

# Use the sci-kit learn stuff
linreg1 = LinearRegression()
linreg1.fit(x, y)

# generate a bunch of x's for plotting
xstep = 0.01
xstepinv = 100
xpoints = np.arange(0.0, 6.0, xstep)

# compute the predicted y's of the x's
ypoints1 = linreg1.predict(np.c_[xpoints])
ypoints3 = linreg3.predict(np.c_[xpoints, xpoints*xpoints, xpoints*xpoints*xpoints])

# Do the plotting
# first draw a line between the plotting x's and their predicted y's 
plt.plot(xpoints, ypoints1, "r-")
plt.plot(xpoints, ypoints3, "g-")
# Then plot the points we predicted from
plt.plot(x, y ,'bo')
# Setup axis and labels
plt.axis([0, 6.0, 0, 6])
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
#plt.title(r'Over fitting?')
plt.show()

print(linreg3.intercept_)
print(linreg3.coef_)
print(linreg3.predict(np.c_[5, 5*5, 5*5*5]))

```
